```python

l_0=[]
l_1=[]
Cost_fuction_0=[]
Cost_fuction_1=[]
it=[]
y_hat_2=[]
def Gradent_disnt(x,y):
    a=0.001
    ite=1000
    theta_0=1
    theta_1=3
    n=len(x)
    
    for i in range(ite):
        it.append(i)
        y_hat=theta_0+theta_1*x
        print("Vector",' ',y_hat)
        y_hat_2.append(y_hat)
        y_hat_1=np.subtract(y,y_hat)
        print("Vector_(h(x))",' ',y_hat_1)
        gradint_theata_0=(1/n)*(np.abs(np.sum(y_hat_1)))
        Cost_fuction_0.append(gradint_theata_0)
        print(np.abs(np.sum(y_hat_1*x)))
        gradint_theata_1=(1/n)*(np.abs(np.sum(y_hat_1*x)))
        Cost_fuction_1.append(gradint_theata_1)
        print("gradint_theata_1",gradint_theata_1)
        theta_0=theta_0-(a*gradint_theata_0)
        theta_1=theta_1-(a*gradint_theata_0)
        l_0.append(theta_0)
        l_1.append(theta_1)
        print("theta_0",theta_0)
        print("theta_1",theta_1)
        print('W',theta_0,'b',theta_1,'cost_fuction',gradint_theata_0)
x = np.array([100,95,90,80,80,70,70,60])
y = np.array([300,285,270,240,235,200,205,180])             


Gradent_disnt(x,y)  
```
